<h3> CLIMBING STAIRS </h3>
<br> It takes n steps to rech the top of a staircase.
If we can take 1 step or 2 step at every point.<br>
Find number of distinct ways to reach the top <br>

The problem is a classical recursive problem..(FIB)<br>
Where the recursive realtion pans out to be <br>

climbStaris(n) = climbStair(n-1) + climbStairs(n-2)<br>
Can reduce it to using only two variables also

```python
def climbStairs(self, n):
        """
        n steps needed to reach the top..
        choice: climb one step or climb two steps
        HOW MANY WAYS ARE THERE TO REACH THE TOP?
        0--> 0
        1--> 1
        2--> 1+1 or 2 :     2
        3--> 1+1+1 or 2+1 or 1+2:   3
        4--> 1+1+1+1 or 2+1+1 or 1+2+1 or 1+1+2 or 2+2:  5
        dp(4) = dp(4-2) + dp(4-1) = 2 + 3 = 5
        
        
        """
        if n<4:
            return n
        dp = [0]*(n+1)
        dp[1] = 1
        dp[2] = 2
        for i in range(3,n+1):
            dp[i] = dp[i-1] + dp[i-2]
        return dp[-1]
```

<h3> UNIQUE PATHS </h3><br>
Find possible paths to reach from 0,0 to n-1,n-1 in a m*n grid

This problem is clasical dp of finding paths, as we are allowed only moving down or right that means we can come only from above or from the left!<br><br>
If we use dp[i][j] for the number of unique paths to arrive at the point (i, j),<br> then the state equation is <br>dp[i][j] = dp[i][j - 1] + dp[i - 1][j].<br>  Base cases dp[0][j] = dp[i][0] = 1 for all valid i and j.<br><br>
We can also reduce it to using a single column.


```python
def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        we are at 0,0
        at each point we can make two choices, 
        either come from top or come from left
        """
        dp = [[1 for i in range(n)]for i in range(m)]
        
        for i in range(1,m):
            for j in range(1, n):
                dp[i][j] = dp[i-1][j] + dp[i][j-1]
        return dp[-1][-1]
           
```

 However, you may have noticed that each time when we update dp[i][j], we only need dp[i - 1][j] (at the previous row) and dp[i][j - 1] (at the current row).

<h3> UNIQUE PATHS II</h3>
Similar to unique paths 1, but now we have obstacles in the matix denoted by 1s

```python
def uniquePathsWithObstacles(self, arr):
        """
        :type obstacleGrid: List[List[int]]
        :rtype: int
        """
        m = len(arr)
        n = len(arr[0])
        
        if(arr[0][0]==1):
            return 0 # No path is posssible!
        
        # else it is obvious that this can  be reached in 1 way only
        arr[0][0] = 1
        
        # Now we fill the first row and the first column
        # If the column I am at is = 0 (no obstacle)
        # If the column before me is = 1 (was valid)
        
        # COLUMNS
        for i in range(1,m):
            # if it is allowed to step on me then count me else its going to be 0 any way
            # now in case you can step on me and you are coming from a place where it was allowed to step (you earned 1 there)
            # then make me 1 too! else let it be false i.e. 0
            arr[i][0] = int(arr[i][0] == 0 and arr[i-1][0] == 1)
        
        # ROWS
        for j in range(1, n):
            arr[0][j] = int(arr[0][j] == 0 and arr[0][j-1] == 1)
            
        for i in range(1,m):
            for j in range(1, n):
                if arr[i][j] == 0:
                    # if it is allowed to step on me, cool! come from top and left sum them
                    arr[i][j] = arr[i-1][j] + arr[i][j-1]
                else:
                    # if not, sad, just say i cant contribute anythin
                    arr[i][j] = 0
                    
        return arr[m-1][n-1]
                
```

<h3> TARGET SUM</h3><br>
We are given a set of numbers, we have to see if we can make a total equal to the given target, making use of all these number, we have a choice of either assigning it to a -ve or a +ve value <br><br>
Same as knapsack the only variation is that we have to USE ALL THE GIVEN NUMBERS! <br>
Choice is either make it positive or make it negative <br>


```python
def findTargetSumWays(self, nums, S):
        index = len(nums) - 1
        curr_sum = 0
        self.memo = {}
        return self.dp(nums, S, index, curr_sum)
        
def dp(self, nums, target, index, curr_sum):
        if (index, curr_sum) in self.memo:
            return self.memo[(index, curr_sum)]
        
        # Base cases
        if index < 0 and curr_sum == target:
            return 1
        if index < 0:
            return 0 
        
        # Either take the number as positive or take it as a negative
        positive = self.dp(nums, target, index-1, curr_sum + nums[index])
        negative = self.dp(nums, target, index-1, curr_sum + -nums[index])
        
        # Upto this index the sum can be formed in the 2 above discused ways
        self.memo[(index, curr_sum)] = positive + negative
        return self.memo[(index, curr_sum)]
```

This problem can also be reduced to, counting ways to form a given sum (target//2) <br> Because if target//2 can be formed then there is surely a way to from the the the other half<br>
                  sum(P) - sum(N) = target<br>
sum(P) + sum(N) + sum(P) - sum(N) = target + sum(P) + sum(N)<br>
                       2 * sum(P) = target + sum(nums)<br>
<br>So the original problem has been converted to a subset sum problem as follows:<br>
Find a subset P of nums such that sum(P) = (target + sum(nums)) / 2



```python
def findTargetSumWays(self, nums, S):
        """
        :type nums: List[int]
        :type S: int
        :rtype: int
        """
        total = sum(nums)
        n = len(nums)
        if (S>total or (S+total)%2!=0):
            return 0
        def countSubsets(set, n, target):
            dp = [[0 for i in range(target+1)] for j in range(n+1)]
            for i in range(n+1):
                dp[i][0] = 1
            for i in range(1,n+1):
                for j in range(1,target+1):
                    if(j>=set[i-1]):
                        dp[i][j] = dp[i-1][j] + dp[i-1][j- set[i-1]]
                    else:
                        dp[i][j] = dp[i-1][j]

            return dp[n][target]
        return countSubsets(nums, n, (S+total)//2)
    
```

<h3> NUMBER OF DICE ROLLS WITH TARGET SUM </h3>
You have d dice, and each die has f faces numbered 1, 2, ..., f.<br>
Return the number of possible ways (out of fd total ways) modulo 10^9 + 7 to roll the dice so the sum of the face up numbers equals target.

```python
class Solution(object):
    def numRollsToTarget(self, d, f, target):
        """
        :type d: int
        :type f: int
        :type target: int
        :rtype: int
        
        d dice in hand
        f faces on each dice
        target needed to make: given
        
        SIMILAR TO COIN CHANGE, amount == target
        and the other thing, coins == [1,f] (They are not infinte but limited by---> d)
        
        2 dice are available.. they are 6 faced
        Target needed: 7
        7 ---> [1,6]
        7 ---> [2,5]
        7 ---> [3,4]
        7 ---> [4,3]
        7 ---> [5,2]
        7 ---> [6,1]
        
        7 comes.... (in outer loop)
        7 now will choose from the options available...
        if 7-val>=0 we will try it, else does not make sense
        [1,2,3,4,5,6]....
        7 will do it f times, the moment 7 sees its not possible to make a choice now, it 
        will say it cant make it!
        
        
        Target needed: 6
        6 --> [1,5]
        6 --> [2,4]
        6 --> [3,3]
        6 --> [4,2]
        6 --> [5,1]
        """
        dp = [[0 for i in xrange(target+1)]for i in xrange(d)]
        dp[0][0] = 0 # WAYS OF USING THE FIRST DICE TO MAKE A TARGET OF 0
        
        for i in xrange(1, min(f,target)+1):
            dp[0][i] = 1 # for the first dice, it can form any sum upto the faces value it has only 
            # and that also only in one way!
        
        for z in xrange(1,d): # for the state of dp table
            for i in xrange(1, target+1): # we will sum up to target startin from 1
                for x in xrange(1, f+1): # pickup the numbers(faces)
                    if i-x>=0: # if the face is capable enough
                        dp[z][i]+= dp[z-1][i-x]
        return dp[-1][-1] % 1000000007
        
```

<h3> COMBINATION SUM </h3>
<br>Given an integer array with all positive numbers and no duplicates, find the number of possible combinations that add up to a positive integer target.

```python
class Solution(object):
    def combinationSum4(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        [1, 2, 3]
        How many total combinations of numbers are there, that can sum to target?
        Unbounded knapsack.
        Coz we have to reach the target, using any number of given nums...
        We will build up our solution
        dp(i) tells ways of making a target of i value
        dp = [0] *(len(target)+1)
        dp[0] = 1 (no ways of making a change of 0)
        for i in range(1, len(target)+1):
            for j in range(nums):
                if i-j>=0: # j is capable of something
                    dp[i] = dp[i] + dp[i-j]
        return dp[-1]
        dp[0] = 1
        dp[1] = 1
        dp[2] = 2
        dp[3] = 2+1+1 = 4
        dp[4] = 4+2+1 = 7
                
        """
        dp = [0] *(target+1)
        dp[0] = 1 
        for i in xrange(1, target+1):
            for num in nums:
                if i-num>=0:
                    dp[i] += dp[i-num]
                 
        return dp[-1]
        
```

<h3> PARTITION EQUAL SUBSET SUM </h3>


```python
class Solution(object):
    def canPartition(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        [1, 5, 11, 5]
        Can I divide this set into 2 subsets
        such that sum of both the subsets is the same!
        [1, 5, 5] , [11] TRUE
        [1], [5,11,5]
        [1,5] , [11,5]
        .......................
        
        At each point in this array,I have a choice.
        Include this guy in ONE, Include this guy in TWO...
        
        1 5 11 5
        1 2 3 5
        
        We need to find, are there some elements in this array that sum to sum(nums)//2
        Beacause that surely means that there are some guys that will give me the remaining sum!
        1 5 11 5---> 22.... find if 11 is possible out of these guys?
        
        """
        
        target = sum(nums)
        if target%2!=0:
            return False
        n = len(nums)
        target = target//2
        dp = [[False for _ in xrange(target+1)]for _ in xrange(n+1)]
        
        for i in range(n+1):
            dp[i][0] = True # It is always possible to make a sum of 0
            
        for i in range(n+1):
            for j in range(target+1):
                if j>=nums[i-1]:
                    dp[i][j] = dp[i-1][j] or dp[i-1][j-nums[i-1]]
                else:
                    dp[i][j] = dp[i-1][j]
        return dp[-1][-1]
        
        
```

<h3> PARTITON TO K EQUAL SUBSETS (WITH SAME SUM)

```python
class Solution(object):
    def canPartitionKSubsets(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        [4, 3, 2, 3, 5, 2, 1]
        [1, 2, 2, 3, 3, 4, 5]
        k = 4.. Can I divide the array into 4 subsets , sum of all must be equal
        
        Basically, we say we have only sacks, we need to fill each bag to the same level
        [   5  1,4  2,3   2,3 ]
        
        For each number I have a choice.
        Put it in the first/second/third/fourth
        
        Take mark of the current maximum at all times
        Greeedy method: Try to balance out things with each step
        Backtracking method: Try out all the possibilities and choose the one which fits
        We back track from the cases that make things impossible for us.
        First check if the sum can be partitioned in K subsets - if its not possible then just return false
        Divide the total sum into subset sum = sum / k
        Check if that subset sum exists first
        If that subset sum exists then check if there are k such sums present
        Use a visited array to keep track of numbers checked to avoid overflow
        [4,1,4,3]
        3 sacks
        [4+1+4] [3] []
        """
        total = sum(nums)
        if k == 0 or total%k!=0:
            return False
        total = total//k
        
         
        visited = [False]*len(nums)
        return self.helper(nums, total, 0, k, 0, visited)
    
    def helper(self, nums, total, currSum, k, start, visited):
        if k == 1:
            return True
        if currSum == total:
            return self.helper(nums, total, 0, k-1, 0, visited)
        for i in range(start, len(nums)):
            if not visited[i] and currSum+nums[i]<=total:
                visited[i] = True
                if self.helper(nums, total, currSum+ nums[i], k, i+1, visited):
                    return True
                visited[i] = False
        return False
    
```
